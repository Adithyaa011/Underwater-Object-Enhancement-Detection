```python
import os
from PIL import Image
import matplotlib.pyplot as plt

result_folder = "/content/drive/MyDrive/FYP_APRIL/Resolution_enhancer/exps/quickstart_Enh/test_results"

# Get a list of all files in the folder
files = os.listdir(result_folder)

# Filter out only the image files (assuming they have common image extensions)
image_files = [f for f in files if f.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp'))]

# Loop through the image files and print them
for image_file in image_files:
    img_path = os.path.join(result_folder, image_file)
    img = Image.open(img_path)
    plt.imshow(img)
    plt.title(image_file)
    plt.axis('off')  # Hide axis
    plt.show()
```

```python
import os
import cv2

# Directory containing images
directory = '/content/drive/MyDrive/FYP_APRIL/Resolution_enhancer/exps/quickstart_Enh/test_results'

# Iterate through files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Check if file is an image
        # Load image
        image_path = os.path.join(directory, filename)
        image = cv2.imread(image_path)

        # Get image dimensions
        height, width, _ = image.shape

        # Crop and resize right half
        right_half = image[:, width//2:, :]
        resized_image = cv2.resize(right_half, (256, 256))

        # Save processed image
        output_path = os.path.join(directory,  filename)
        cv2.imwrite(output_path, resized_image)

print("Processing complete.")
```

```python
import os
import shutil

source_dir = '/content/drive/MyDrive/FYP_APRIL/Resolution_enhancer/exps/quickstart_Enh/test_results'
destination_dir = '/content/drive/MyDrive/FYP_APRIL/ImageRestoration/hq_demo/data/datasets/gts/face'

# Check if both source and destination directories exist
if os.path.exists(source_dir) and os.path.exists(destination_dir):
    # Get the list of files in the source directory
    files = os.listdir(source_dir)

    # Iterate over each file in the source directory
    for file_name in files:
        # Construct the source and destination file paths
        source_file = os.path.join(source_dir, file_name)
        destination_file = os.path.join(destination_dir, file_name)

        # Move the file from the source to the destination directory
        shutil.move(source_file, destination_file)

    print("Images moved successfully.")
else:
    print("Source or destination directory does not exist.")
```
